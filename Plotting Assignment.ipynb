```python
# YOUR CODE HERE
# extract the co2 data
co2_data = fuel_econ_data.co2
# skew for the co2 data
skew = stats.skew(co2_data)
print(skew)
```

```python
# YOUR CODE HERE
bin_width = 20
# configure the plot and plot the histogram
plt.figure(figsize = [8,3])
plt.hist(co2_data, bins = bin_width, color = 'darkgreen', alpha = 0.7)
plt.xlabel('CO2 (g/km)')
plt.ylabel('Frequency')
plt.title('Carbon Dioxide Emissions')
```

```python
# YOUR CODE HERE
fuel_data = fuel_econ_data.fuelType
bin_width = 15
plt.hist(fuel_data, bins = bin_width, color = 'darkblue')
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.title('Fuel Type Distribution')
```

```python
high_way_data = fuel_econ_data.UHighway
bin_width = 15
plt.hist(high_way_data, bins = bin_width, color = 'y')
plt.xlabel('Highway Fuel Economy')
plt.ylabel('Frequency')
plt.title('High way Distribution')
```

```python
# Define custom colormap with grey and blue
cmap = sns.diverging_palette(220, 20, as_cmap=True)

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data=fuel_econ_data[['disp_liters', 'co2']].corr(), annot=True, cmap=cmap, linewidths=.5)
plt.xlabel('Engine Displacement (Liters)')
plt.ylabel('CO2 Emissions (g/km)')
plt.title('Correlation Heatmap: Engine Displacement vs. Carbon Dioxide Emissions')
plt.show()
```
